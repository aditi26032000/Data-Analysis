# 1. Data Cleaning and Preprocessing 

```python
# No null values are found in dataset
# Start analysing columns with string data type
data['CarName'].unique()
```

```python
# Seperate company name from car name 
data['CarName']= data['CarName'].str.split(" ", n = 1, expand = True) 
```

```python
data['CarName'].unique()
```

```python
# corect spelling errors and convert in lower case all carname 
data['CarName']= data['CarName'].str.lower()

def change_carname(x,y):
    data['CarName'].replace(x,y,inplace=True)

change_carname('maxda','mazda')
change_carname('porcshce','porsche')
change_carname('toyouta','toyota')
change_carname('vokswagen','volkswagen')
change_carname('vw','volkswagen')
```

```python
data['CarName'].unique()
```

```python
 
print(data['fueltype'].unique())
print(data['aspiration'].unique())
print(data['doornumber'].unique())
print(data['carbody'].unique())
print(data['drivewheel'].unique())
print(data['enginelocation'].unique())
print(data['fuelsystem'].unique())
```

```python
#convert string data to numerical
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
```

```python
data['fueltype']= le.fit_transform(data['fueltype']) 
data['aspiration']= le.fit_transform(data['aspiration']) 
data['doornumber']=le.fit_transform(data['doornumber']) 
data['carbody'] = le.fit_transform(data['carbody']) 
data['drivewheel'] = le.fit_transform(data['drivewheel']) 
data['enginelocation'] = le.fit_transform(data['enginelocation']) 
data['fuelsystem'] = le.fit_transform(data['fuelsystem'] ) 
data['CarName'] = le.fit_transform(data['CarName'] )
data['enginetype'] = le.fit_transform(data['enginetype'] ) 
data['cylindernumber'] = le.fit_transform(data['cylindernumber'] ) 
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['wheelbase', 'curbweight', 'enginesize', 'boreratio', 'horsepower','carlength','carwidth','price']
data[num_vars] = scaler.fit_transform(data[num_vars])
```

```python
# divide input and output
Y = data['price']
data.drop(['price'],axis=1,inplace =True)
```

```python
# test train and split
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(data,Y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# import libraries for model building
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RandomizedLasso
```

```python
#Correlation using heatmap
plt.figure(figsize = (30, 25))
sns.heatmap(X_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

# 3. Model Building

```python
model = LinearRegression()
model.fit(X_train,Y_train)

model_rfe = RFE(model, 10)
model_rfe.fit(X_train, Y_train)
print("Num Features: %d" % model_rfe.n_features_)
print("Selected Features: %s" % model_rfe.support_)
print("Feature Ranking: %s" % model_rfe.ranking_)
```

```python
# Plot the errors
Y_predict = model.predict(X_test)
fig = plt.figure()
sns.distplot((Y_predict - Y_test), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                 
plt.xlabel('Errors', fontsize = 18)   
```

```python
# Plotting Y_test and Y_predict 
fig = plt.figure()
plt.scatter(Y_test,Y_predict)
fig.suptitle('y=Y_test vs Y_predict', fontsize=20)              
plt.xlabel('Y_test', fontsize=18)                          
plt.ylabel('Y_predict', fontsize=16)  
```

```python
# checking for error
from sklearn import metrics
metrics.mean_squared_error(Y_test,Y_predict)
```
